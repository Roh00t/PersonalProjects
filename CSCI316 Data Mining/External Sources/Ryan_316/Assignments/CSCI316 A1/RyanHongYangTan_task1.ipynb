Creating of data frame and cleaning of data

```python
import pandas as pd
import numpy as np
from google.colab import drive
from scipy.stats import zscore
drive.mount("/content/drive")

data = pd.read_csv('/content/drive/My Drive/drug200_v2.csv')
drugDF = pd.DataFrame(data,columns=['Age','Sex','BP','Cholesterol','Na_to_K','Drug'])

# Identifying the number of missing values per column
print(drugDF.isnull().sum())

# Identifying the rows with missing values
incompleteRows = drugDF[drugDF.isnull().any(axis=1)]
print('\n', incompleteRows, '\n')

# Calculating the mode value of BP (due to categorical data) and getting the first value
BPMode = drugDF['BP'].mode()[0]
# Cleaning the data set by filling null value of BP column with the mode value of dataset
drugDF['BP'] = drugDF['BP'].fillna(BPMode)
# Checking for any null value
print(drugDF.isnull().sum())
```

Z-score normalisation and getting of mean and variance of the normalised value

```python
mean_value = drugDF['Na_to_K'].mean()
std_value = drugDF['Na_to_K'].std()

# Getting the Z Score
# Na_to_K_ZScore = zscore(drugDF['Na_to_K'])
Na_to_K_ZScore = (drugDF['Na_to_K'] - mean_value) / std_value
# Appending Na to K Z Score column into the dataframe
drugDF['Na_to_K_ZScore'] = Na_to_K_ZScore
print(drugDF)

# Getting the mean and variance
normalizedMean = drugDF['Na_to_K_ZScore'].mean()
normalizedVariance = drugDF['Na_to_K_ZScore'].var()

print('\nMean of normalized value: ', normalizedMean)
print('Variance of normalized value: ', normalizedVariance)
```

Creation and distribution of Na_to_K value evenly among 4 bins

```python
# Distribute among 4 bins evenly
drugDF['Na_to_K_bin'] = pd.qcut(drugDF['Na_to_K'], 4, labels=[1, 2, 3, 4])

# Checking if the bins are (approximately) equivilant in size
print('Size of bin 1: ', len(drugDF[drugDF['Na_to_K_bin'] == 1]))
print('Size of bin 2: ', len(drugDF[drugDF['Na_to_K_bin'] == 2]))
print('Size of bin 3: ', len(drugDF[drugDF['Na_to_K_bin'] == 3]))
print('Size of bin 4: ', len(drugDF[drugDF['Na_to_K_bin'] == 4]))

print('\n', drugDF)
```

Applying one-hot-encoding on Cholesterol values

```python
# Get the different OHE values
CholesterolOHE = pd.get_dummies(drugDF['Cholesterol'], prefix = 'CholesterolOHE')
print(CholesterolOHE)

drugDF = pd.concat([drugDF, CholesterolOHE], axis = 1)
print(drugDF)
```
